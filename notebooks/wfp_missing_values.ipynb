```python
import pandas as pd
from sklearn.impute import KNNImputer
import numpy as np
```

```python
df = pd.read_csv('/Users/pavelpavlov/Desktop/gaussian_peppers/food_diets_eda/raw_data/world-food-prices/wfp_food_prices_global_2017.csv')
df
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['usdprice'] = pd.to_numeric(df['usdprice'], errors='coerce')
```

```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
```

```python
# Initialize KNN imputer
imputer = KNNImputer(n_neighbors=3, weights='distance')
```

```python
# Impute only numeric columns
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])

print(df)
```
