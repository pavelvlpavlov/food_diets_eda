```python
gdp_clean = gdp.dropna()
```

```python
gdp_clean
```

```python
# Dropping unnecessary columns
gdp_clean2 = gdp_clean.drop(columns=['Series Name', 'Series Code'])
```

```python
# Changing the year columns from wide to long format
gdp_long = gdp_clean2.melt(id_vars=['Country Name', 'Country Code'], var_name='year', value_name='gdp_per_capita')
gdp_long
```

```python
gdp_long['gdp_per_capita'] = pd.to_numeric(gdp_long['gdp_per_capita'], errors='coerce')
gdp_long['year'] = gdp_long['year'].astype(str).str.extract(r'(\d{4})').astype(int)
```

```python
gdp_long
```

```python
gdp_long_clean = gdp_long.dropna(subset=['gdp_per_capita'])
gdp_long_clean
```
