```python
import pandas as pd

wfp_df = pd.read_csv('raw_data/world-food-prices/wfp_food_prices_global_2017.csv')
```

```python
wfp_df
```

```python
wfp_df = wfp_df[['countryiso3', 'category', 'usdprice']]
wfp_df
```

```python
wfp_df = wfp_df.iloc[1:]
```

```python
wfp_df
```

```python
wfp_df.info()
```

```python
wfp_df = wfp_df.rename(columns={'countryiso3': 'country'})
wfp_df
```

```python
wfp_df
```

```python
wfp_df[wfp_df[['country', 'category', 'usdprice']].isnull().any(axis=1)]
```

```python
wfp_df['usdprice'] = wfp_df['usdprice'].astype(float)
```

```python
wfp_df.info()
```

```python
agg_df = wfp_df.groupby(['country', 'category'])['usdprice'].agg(sumPrice='sum', meanPrice='mean').reset_index()
agg_df
```

```python
agg_df
```

```python

agg_df.to_csv('cleaned_data/wfp_food_prices_cleaned_2017.csv', index=False)
```
